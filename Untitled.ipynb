```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
df = pd.read_csv('Measurements.csv')

plt.scatter(df['Height (in cm)'],df['Weight (in kg)'])
plt.xlabel('Height (in cm)')
plt.ylabel('Weight (in kg)')
plt.title('Measurements Height vs Weight')
plt.show()
```

```python
correlation_coefficient, p_value = stats.pearsonr(df['Height (in cm)'],df['Weight (in kg)'])

print(f'the correlation coefficient of weight and heigh is {correlation_coefficient:.4f}')
```

```python
import pandas as pd
from scipy import stats

# Step 1: Import the required libraries

# Step 2: Load the CSV file into a pandas DataFrame
data = pd.read_csv('iris.data.csv')

# Perform linear regression
regression_result = stats.linregress(data['sepal length'], data['sepal width'])

# Extract regression coefficients
intercept = regression_result.intercept
slope = regression_result.slope
coefficient_of_determination = regression_result.rvalue**2

print(f"Intercept: {intercept:.4f}")
print(f"Slope: {slope:.4f}")
print(f"Coefficient of Determination: {coefficient_of_determination:.4f}")
print()

# Filter data for species 'setosa'
data_setosa = data[data['species'] == 'setosa']

setosa_res = stats.linregress(data_setosa['sepal length'], data_setosa['sepal width'])

setosa_int = setosa_res.intercept
setosa_slope = setosa_res.slope
coefficient_of_determination_setosa = setosa_res.rvalue**2

print(f"Intercept for Setosa: {setosa_int:.4f}")
print(f"Slope for Setosa: {setosa_slope:.4f}")
print(f"Coefficient of Determination for Setosa: {coefficient_of_determination_setosa:.4f}")
print()
import pandas as pd
from scipy import stats

# Step 1: Import the required libraries

# Step 2: Load the CSV file into a pandas DataFrame
data = pd.read_csv('iris.data.csv')

# Filter data for species 'versicolor'
data_versicolor = data[data['species'] == 'versicolor']

# Perform linear regression for 'versicolor'
versicolor_res = stats.linregress(data_versicolor['sepal width'], data_versicolor['sepal length'])

versicolor_int = versicolor_res.intercept
versicolor_slope = versicolor_res.slope
coefficient_of_determination_versicolor = versicolor_res.rvalue**2

print(f"Intercept for Versicolor: {versicolor_int:.4f}")
print(f"Slope for Versicolor: {versicolor_slope:.4f}")
print(f"Coefficient of Determination for Versicolor: {coefficient_of_determination_versicolor:.4f}")
print()

# Filter data for species 'virginica'
data_virginica = data[data['species'] == 'virginica']

# Perform linear regression for 'virginica'
virginica_res = stats.linregress(data_virginica['sepal width'], data_virginica['sepal length'])

virginica_int = virginica_res.intercept
virginica_slope = virginica_res.slope
coefficient_of_determination_virginica = virginica_res.rvalue**2

print(f"Intercept for Virginica: {virginica_int:.4f}")
print(f"Slope for Virginica: {virginica_slope:.4f}")
print(f"Coefficient of Determination for Virginica: {coefficient_of_determination_virginica:.4f}")
print()

```
